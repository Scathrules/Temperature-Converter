```python
import tkinter as tk
from tkinter import ttk
%gui tk

def temperature_converator():
    
    temperature_value = ent_temperature.get()
    
    if (cb_temp_init.get()=="") or (cb_temp_result.get()==""):
        pass
    
    elif (cb_temp_init.get()==cb_temp_result.get()):
        lbl_result["text"] = f"{round(float(temperature_value),2)}"
    
    elif (cb_temp_init.get()=="\N{DEGREE FAHRENHEIT}") and (cb_temp_result.get()=="\N{DEGREE CELSIUS}"):
        fah_to_cel = (5/9) * (float(temperature_value) - 32)
        lbl_result["text"] = f"{round(fah_to_cel,2)}"
    
    elif (cb_temp_init.get()=="\N{DEGREE FAHRENHEIT}") and (cb_temp_result.get()=="K"):
        fah_to_kel = ((float(temperature_value) - 32) * (5/9)) + 273.15
        lbl_result["text"] = f"{round(fah_to_kel,2)}"
    
    elif (cb_temp_init.get()=="\N{DEGREE CELSIUS}") and (cb_temp_result.get()=="\N{DEGREE FAHRENHEIT}"):
        cel_to_fah = ((9/5) * float(temperature_value)) + 32
        lbl_result["text"] = f"{round(cel_to_fah,2)}"
    
    elif (cb_temp_init.get()=="\N{DEGREE CELSIUS}") and (cb_temp_result.get()=="K"):
        cel_to_kel = float(temperature_value) + 273.15
        lbl_result["text"] = f"{round(cel_to_kel,2)}"
    
    elif (cb_temp_init.get()=="K") and (cb_temp_result.get()=="\N{DEGREE FAHRENHEIT}"):
        kel_to_fah = ((9/5) * (float(temperature_value) - 273.15)) + 32
        lbl_result["text"] = f"{round(kel_to_fah,2)}"
    
    elif (cb_temp_init.get()=="K") and (cb_temp_result.get()=="\N{DEGREE CELSIUS}"):
        kel_to_cel = float(temperature_value) - 273.15
        lbl_result["text"] = f"{round(kel_to_cel,2)}"
        

temp_units = ["\N{DEGREE FAHRENHEIT}","\N{DEGREE CELSIUS}", "K"]
        
window = tk.Tk()
window.title("Temperature Converter")
window.resizable(width=False, height=False)

frm_entry = tk.Frame(master=window)
ent_temperature = tk.Entry(master=frm_entry, width=10)
cb_temp_init = ttk.Combobox(master=frm_entry, values=temp_units, width=5,
                            state="readonly")

ent_temperature.grid(row=0, column=0, sticky="e")
cb_temp_init.grid(row=0, column=1, sticky="w")

btn_convert = tk.Button(master=window, 
                        text="\N{RIGHTWARDS BLACK ARROW}",
                       command=temperature_converator)
cb_temp_result = ttk.Combobox(master=window, values=temp_units, width=5,
                             state="readonly")
lbl_result = tk.Label(master=window)

frm_entry.grid(row=0, column=0, padx=10)
btn_convert.grid(row=0, column=1, pady=10)
lbl_result.grid(row=0, column=2,padx=10)
cb_temp_result.grid(row=0, column=3, padx=10)

window.mainloop()


```
