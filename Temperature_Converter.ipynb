```python
import tkinter as tk
from tkinter import ttk
%gui tk 

#Using a list to store temperature units for convenience
temp_units = ["\N{DEGREE FAHRENHEIT}","\N{DEGREE CELSIUS}", "K"]

def temperature_converter():
    """Button command function for converting between temperature units"""
    
    #checking if the entry can be converted to a float
    try:
        temperature_value = float(ent_temperature.get())
    except ValueError:
    #if the entry is anything but a number (float), the button will not run so as to avoid an exception error
        pass
    else:
        
        #condition: if the initial unit and the desired unit is empty, the button will not run
        if (cb_temp_init.get()=="") or (cb_temp_result.get()==""):
            pass
        
        #condition: if the initial unit and the desired unit is the same, the resulting temperature is the same
        #as the initial temperature
        elif (cb_temp_init.get()==cb_temp_result.get()):
            lbl_result["text"] = f"{round(temperature_value,2)}"
        
        ######################################################################
        #fahrenheit to celsius
        elif (cb_temp_init.get()==temp_units[0]) and (cb_temp_result.get()==temp_units[1]):
            fah_to_cel = (5/9) * (temperature_value - 32) #conversion formula
            lbl_result["text"] = f"{round(fah_to_cel,2)}"
        
        #fahrenheit to kelvin
        elif (cb_temp_init.get()==temp_units[0]) and (cb_temp_result.get()==temp_units[2]):
            fah_to_kel = ((temperature_value - 32) * (5/9)) + 273.15 #conversion formula
            lbl_result["text"] = f"{round(fah_to_kel,2)}"
        
        #celsius to fahrenheit
        elif (cb_temp_init.get()==temp_units[1]) and (cb_temp_result.get()==temp_units[0]):
            cel_to_fah = ((9/5) * temperature_value) + 32 #conversion formula
            lbl_result["text"] = f"{round(cel_to_fah,2)}"
        
        #celsius to kelvin
        elif (cb_temp_init.get()==temp_units[1]) and (cb_temp_result.get()==temp_units[2]):
            cel_to_kel = temperature_value + 273.15 #conversion formula
            lbl_result["text"] = f"{round(cel_to_kel,2)}"
        
        #kelvin to fahrenheit
        elif (cb_temp_init.get()==temp_units[2]) and (cb_temp_result.get()==temp_units[0]):
            kel_to_fah = ((9/5) * (temperature_value - 273.15)) + 32 #conversion formula
            lbl_result["text"] = f"{round(kel_to_fah,2)}" 
        
        #kelvin to celsius
        elif (cb_temp_init.get()==temp_units[2]) and (cb_temp_result.get()==temp_units[1]):
            kel_to_cel = temperature_value - 273.15 #conversion formula
            lbl_result["text"] = f"{round(kel_to_cel,2)}"

#create main window with custom title     
window = tk.Tk()
window.title("Temperature Converter")
#set window geometry
window.resizable(width=False, height=False)

#create frame to hold entry and combobox widgets
frm_entry = tk.Frame(master=window)
#create entry for inputting temperature
ent_temperature = tk.Entry(master=frm_entry, width=10)
#create combobox for initial temperature with readonly values from temp_units
cb_temp_init = ttk.Combobox(master=frm_entry, values=temp_units, width=5, state="readonly")
#create conversion button with a right black arrow as its label
btn_convert = tk.Button(master=window, text="\N{RIGHTWARDS BLACK ARROW}", command=temperature_converter)
#create combobox for desired temperature with readonly values from temp_units
cb_temp_result = ttk.Combobox(master=window, values=temp_units, width=5, state="readonly")
#create label for converted temperature
lbl_result = tk.Label(master=window)

#setting geometry of the widgets
ent_temperature.grid(row=0, column=0, sticky="e")
cb_temp_init.grid(row=0, column=1, sticky="w")
frm_entry.grid(row=0, column=0, padx=10)
btn_convert.grid(row=0, column=1, pady=10)
lbl_result.grid(row=0, column=2,padx=10)
cb_temp_result.grid(row=0, column=3, padx=10)

window.mainloop()

```
